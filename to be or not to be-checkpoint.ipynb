```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset= pd.read_csv('https://raw.githubusercontent.com/aminshojaei/To-be-or-not-to-be/master/Shakespeare_data.csv')
dataset.head()
```

```python
dataset.dropna(how='any', inplace= True)
players_unique = dataset.Player.unique()
play_unique = dataset.Play.unique()
print('Number of unique player is: ',len(players_unique))
print('Number of unique play is: ',len(play_unique))
```

- Now I wanna show that how many players play in specific play. For doing this following code are provided.

```python
number_of_player = dataset.groupby(['Play'])['Player'].nunique().sort_values(ascending= False).to_frame()
S=pd.Series(list(np.arange(0,len(play_unique),1)))
number_of_player = dataset.groupby(['Play'])['Player'].nunique().sort_values(ascending= False).to_frame()
number_of_player['Play'] = number_of_player.index.tolist()
number_of_player=number_of_player.set_index([S])
number_of_player.head(10)
```

- Now I want to change the order of the columns and change the name of them to make the dataset more clear.

```python
number_of_player.rename(columns={'Player':'number of player'}, inplace=True)
cols=number_of_player.columns.tolist()
cols=cols[-1:]+cols[:-1]
number_of_player = number_of_player[cols]
number_of_player.head(10)
```

```python
plt.figure(figsize=(15,10))
sns.set(style='whitegrid')
ax = sns.barplot(x='number of player',y='Play',data=number_of_player)
ax.set(xlabel='Number of Player', ylabel='Play')
plt.show()
```

```python
acts = []
scenes = []
lines = []
for row in dataset["ActSceneLine"]:
    act = row.split('.')[0]
    acts.append(act)

    scene = row.split('.')[1]
    scenes.append(scene)

    line = row.split('.')[2]
    lines.append(line)
dataset.insert(len(dataset.columns), 'Act', acts, True)
dataset.insert(len(dataset.columns), 'Scene', scenes, True)
dataset.insert(len(dataset.columns), 'Line', lines, True)
dataset = dataset.drop(['ActSceneLine'] , axis = 1)
dataset.head(10)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(sublinear_tf=True, min_df=5, norm='l2', encoding='latin-1', ngram_range=(1, 2), stop_words='english')
features = tfidf.fit_transform(dataset.PlayerLine).toarray()
features.shape
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
dataset['PlayerCode'] = LE.fit_transform(dataset['Player'].astype('str')) #generating PlayerCode column
dataset['PlayCode'] = LE.fit_transform(dataset['Play'].astype('str')) #generating PlayCode column
dataset.tail(10)
```

```python
from sklearn.model_selection import train_test_split
X=dataset[['Act','Scene','Line','PlayCode']].values
Y=dataset['PlayerCode'].values
X_train, X_test, y_train, y_test = train_test_split(X, Y,test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(n_estimators=50)
RF.fit(X_train,y_train)
print('Accuracy of Random Forest classifier on tests set: {:.2f}'.format(RF.score(X_test,y_test)))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB

count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(X_train)
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)

# Run Training
clf = MultinomialNB().fit(X_train_tfidf, y_train)
```
